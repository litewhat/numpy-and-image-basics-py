```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("my_super_new_picture.jpg", cv2.COLOR_BGR2RGB)
# img = cv2.imread("my_super_new_picture.jpg", cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(384, 0), pt2=(510, 150), color=(0, 255, 0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200, 200), pt2=(300, 450), color=(0, 150, 0), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(200, 200), radius=100, color=(255, 0, 120), thickness=2)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(100, 100), radius=50, color=(120, 0, 255), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img,
            text="Paweł",
            org=(200, 200),
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
           fontScale=3,
           color=(255, 255, 255),
           thickness=4,
           lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
